You work as a Data Analyst for a national electronics retail chain. The company has been experiencing inconsistent sales performance across different cities and wants to optimize its sales strategy.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sd = pd.read_csv("/content/Sales_Data.csv")
print(sd.shape)
sd.head(10)
```

```python
visualization of the distribution of product with city

product_city_counts = sd.groupby(['Product', 'City'])['Order ID'].count().unstack()

plt.figure(figsize=(12, 6))
sns.heatmap(product_city_counts, annot=True, fmt='d', cmap='viridis')
plt.title('Distribution of Product by City')
plt.xlabel('City')
plt.ylabel('Product')
plt.show()
```

The data shows that **San Francisco leads in sales across multiple product categories**, followed by Los Angeles and New York City. **USB-C Charging Cables, Lightning Cables, and Wired Headphones are the top-selling items**, with battery packs also in high demand. **Larger appliances like washing machines and dryers have lower sales**, while **tech accessories dominate purchases**. Cities with **higher populations and tech-savvy consumers drive sales**, making them key targets for marketing and inventory strategies. 

```python
visualization of correlation between month and product

# Convert 'Order Date' to datetime objects
sd['Order Date'] = pd.to_datetime(sd['Order Date'])
sd['Month'] = sd['Order Date'].dt.month

# Group by 'Month' and 'Product' and count orders
month_product_counts = sd.groupby(['Month', 'Product'])['Order ID'].count().unstack()

# Create the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(month_product_counts, annot=True, fmt='d', cmap='viridis')
plt.title('Correlation between Month and Product Sales')
plt.xlabel('Product')
plt.ylabel('Month')
plt.show()
```

From the data above, we can see that **December has the highest sales** across most products, likely due to holiday shopping. **Tech accessories and batteries** show strong performance throughout the year, while larger electronics experience fluctuations. Sales trends can help optimize inventory and marketing strategies.
